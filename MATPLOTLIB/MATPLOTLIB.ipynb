```python
import matplotlib.pyplot as plt
import numpy as np

#To render chart in a proper format
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [40,50,45,42,48,43,41]
```

### Plot chart

```python
#Specify tile,x/y axis titles
plt.xlabel('Week')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y, color='red', linewidth=2, linestyle='dotted')
```

```python
#Specify the marker to represent the value
#plt.plot(x,y, '--r+')

#plt.plot(x,y,color="green",marker="D",linestyle="",linewidth=2)
plt.plot(x,y,color="#008400",marker="D",markersize=5,linestyle="",linewidth=2, alpha=0.5)
```

```python
days = [1,2,3,4,5,6,7]
max_temperature = [50,45,46,47,43,48,49]
min_temperature = [40,35,36,37,33,38,39]
average_temperature = [45,40,44,38,40,45,48]

plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days,max_temperature, label="Max")
plt.plot(days,min_temperature, label="Min")
plt.plot(days,average_temperature, label="Avg")

#Best fit place within the chart
plt.legend(loc="best", shadow=True, fontsize="small")

plt.grid()
```

### Bar graph

```python
company = ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90,150,100,120]
profit = [40, 5 , 70, 80]
```

```python
xpos = np.arange(len(company))
print(xpos)
```

```python
plt.xticks(xpos,company)
plt.ylabel("Revenue(Billions)")
plt.title("Stock Values")

plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")
plt.legend()
```

```python
plt.yticks(xpos,company)
plt.ylabel("Revenue(Billions)")
plt.title("Stock Values")

plt.barh(xpos-0.2,revenue,label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")
plt.legend()
```

### Histogram - Bar graph representation where y axis represents the frequency of occurance of an event

```python
blood_sugar_men = [113, 92, 91, 80, 84, 100, 120, 90, 80, 81, 95, 112, 149]
blood_sugar_women = [123, 90, 99, 81, 80, 101, 125, 95, 88, 81, 95, 112, 149]

#plt.hist(blood_sugar, bins=[80, 100, 125, 150], rwidth=0.8, color="g")
plt.xlabel("Sugar Range")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Analysis")
#plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], rwidth=0.8, color=["r","g"], label=["Men","Women"])
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], rwidth=0.8, color=["r","g"], label=["Men","Women"], orientation="horizontal")
plt.legend()
```

### Pie Chart

```python
exp_values = [1400, 500, 400, 100, 200]
exp_labels = ["Home Rent", "Water", "Electricity", "Phone", "Internet"]

plt.pie(exp_values, labels=exp_labels, radius=1.2, autopct="%1.0f%%", shadow=True, explode=[0,0.2,0,0,0], startangle=45)
plt.axis("equal")
plt.show()

#Save the chart to a file
plt.savefig("Piechart.png", bbox_inches="tight", dpi=72, pad_inches=2, transparent=True)
```
